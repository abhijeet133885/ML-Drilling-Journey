```python
import requests
from bs4 import BeautifulSoup
import os
```

```python
url='https://archive.org/details/in.gazette.central.e.2007-12-29.009'
try:
    response=requests.get(url,timeout=20)
    if response.status_code==200:
        print('Successfully retrieve the page')
    else:
        print('Failed to retrieve the webpage,status code:',response.status_code)
except requests.exceptions.Timeout:
    print('Request timeout')
except requests.exceptions.RequestException as e:
    print('Error occurred:',e)
```

```python
soup=BeautifulSoup(response.content,'html.parser')
print('Webpage Title',soup.title.text)
```

```python
document_link = soup.find('a',{'class':'hidden-for-screen-readers'})['href']
print('Document link:',document_link)
```

```python
base_url = 'https://archive.org/details/in.gazette.central.e.2007-12-29.009'
full_url = os.path.join(base_url,document_link)
print('full url:',full_url)
```

```python
/**The backslash (\) you’re seeing comes from using os.path.join(), 
which is meant for file system paths, not URLs. Let me explain.
os.path.join() is designed for Windows paths.
On Windows, it uses backslash (\) as the separator.
It doesn’t know you’re working with a URL → so it inserts \ instead of /.
URLs must always use forward slashes (/), never backslashes
```

```python
document_response= requests.get('https://archive.org/details/in.gazette.central.e.2007-12-29.009/#maincontent')

if document_response.status_code==200:
    with open('document.pdf','wb') as file:
        file.write(document_response.content)
    print('Document Downloaded successfully.')
else:
    print('Failed to download the documnet. Satus code:', document_response.status_code)
    
```
